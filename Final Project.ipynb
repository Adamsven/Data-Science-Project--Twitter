```python
#Using this function to save the tweets into a dataframe with text, date, username, number of favorites and RTs
def extractor(tweets):
    l1,l2,l3,l4,l5=[],[],[],[],[]

    for i in range(len(tweets)):
        for tweet in tweets[i]:
            l1.append(tweet.text)
            l2.append(tweet.date)
            l3.append(tweet.username)
            l4.append(tweet.favorites)
            l5.append(tweet.retweets)
        
    final=pd.DataFrame(list(zip(l1,l2,l3,l4,l5)), 
                       columns =['Tweets', 'Date','User', 'Favs', 'RT'])

    
    return final
```

```python
# Saving the data into dataframes
import pandas as pd

df0 = extractor(trump1)
df1 = extractor(sanders1)
```

```python
from sklearn.base import TransformerMixin 

# This function will clean the text
def clean_text(text):     
    return text.strip().lower()
    
#Custom transformer using spaCy 
class predictors(TransformerMixin):

    def transform(self, X, **transform_params):
        return [clean_text(text) for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

    def get_params(self, deep=True):
        return {}
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Using Tfidf
tfvectorizer = TfidfVectorizer(tokenizer = spacy_tokenizer)
```

```python
# Splitting Data Set
from sklearn.model_selection import train_test_split
```

```python
#Defining X and why
X = data2['Tweets']
ylabels = data2['Classifier']
```

```python
#Splitting into the dataset using an 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X, ylabels, test_size=0.2, random_state=42)
```

```python
X_train.shape
y_train.shape
X_test.shape
y_test.shape
```

```python
#Using the LinearSVC classifier from sklearn
from sklearn.svm import LinearSVC

classifier_SVC = LinearSVC(verbose=True)
```

```python
# Now I want to create the  pipeline to clean, tokenize, vectorize, and classify the data
from sklearn.pipeline import Pipeline

pipe = Pipeline([("cleaner", predictors()),
                 ('vectorizer', tfvectorizer),
                 ('classifier', classifier_SVC)], verbose=True)
```

```python
#Fitting my training data to the model 
pipe.fit(X_train, y_train)
```

```python
#Defining the prediction based on the test data 
sample_prediction = pipe.predict(X_test)
```

```python
#I used this in order to check if my model was predicting both 0s and 1s, so I knew if it was working or not
for (sample,pred) in zip(X_test,sample_prediction):
    print(sample,"Prediction=>",pred)
```

```python
#Evaluating the model
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, sample_prediction))
print("Precision:",metrics.precision_score(y_test, sample_prediction))
print("Recall:",metrics.recall_score(y_test, sample_prediction))
```

Model Accuracy, Precision and Recall are all really good. This tells me that the model seems to be working quite well and I am happy with the performance.
